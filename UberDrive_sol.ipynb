Q17. List the most frequent routes(same START & END points) taken by Uber drivers. (2 points)Â¶

```python
#to find the most frequent routes (same START and STOP points) used by uber drivers,
# first group the data by start & stop combinations/routes, then sum the miles per route, sort in descending order and display.
grouped_uber = uber.groupby(['START*', 'STOP*']).sum()
sorted_uber = grouped_uber.sort_values(by=('MILES*'), ascending = False)
sorted_uber
#Unknown route is most frequent of all, if we dont count it, Cary to Durham is the second most frequent route.
```

Q18. List out all the trips undertaken by the Uber drivers in the month of January.
(date format is Month,Day,Year) (2 points)

```python
#to list out all the trips taken in the month of January by uber drivers
jan = uber[pd.to_datetime(uber['START_DATE*']).dt.month == 1]
jan
```

Q19. Print the miles travelled by the Uber drivers on each day in the month of January. (2 points)

```python
#to print all the miles (sum) travelled by Uber drivers daily in the January
groupbyDay_jan = jan.groupby(pd.to_datetime(jan['START_DATE*']).dt.day).sum()['MILES*']
groupbyDay_jan
```
